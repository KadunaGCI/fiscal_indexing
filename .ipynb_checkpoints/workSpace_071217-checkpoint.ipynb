```python
remarkers = []
for file in files:
    csv = pd.read_csv(os.path.join(DATA_DIR,'EID42186/myCsv',file.replace(".xml.gz",".csv"))).drop(['Id', 'TimeOfArrival','Body'],axis=1)
    for headline in csv['Headline']:
        if(len(headline.split('：')) > 1):
            remarkers.append(headline.split('：')[0].replace('*',''))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(min_df=2).fit(remarkers) # stop_words = ''
bow = vectorizer.transform(remarkers)
```

```python
bow.get_shape()
```

```python
# 綺麗に書き換えられる
freqs = [(word, bow.getcol(idx).sum()) for word, idx in vectorizer.vocabulary_.items()]
order = sorted(freqs, key = lambda x: -x[1]) # tapleのlist
terms = []
nums = []
for a, b in order:
    terms.append(a)
    nums.append(b)
```

```python
import seaborn as sns

sns.distplot(nums[0:50], kde=False, rug=False, bins=10) 
```

```python
# 背景おかしい？
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
fig = plt.figure(figsize=(17,3))
ax = fig.add_subplot(1,1,1)
ax.set_xticklabels(terms[0:40], rotation = 90)
sns.set_style("whitegrid")

font_path = '/usr/share/fonts/truetype/takao-gothic/TakaoPGothic.ttf'
font_prop = FontProperties(fname=font_path)
matplotlib.rcParams['font.family'] = font_prop.get_name()
sns.barplot(x=terms[0:40], y=nums[0:40])

plt.title('2014の発言回数トップ20　（※人物のみ）')
```

```python
# 日銀とかは入っていない
from janome.tokenizer import Tokenizer
import time
start = time.time()

def myTokenizer(text):
    wordsIn=[]
    t = Tokenizer()
    tokens = t.tokenize(text)
    for token in tokens:
        #print(token)
        #if (token.part_of_speech.split(',')[0]=='名詞'):
        if (token.part_of_speech.split(',')[2]=='人名'):
            wordsIn.append(token.surface)
    return wordsIn

vectorizer = CountVectorizer(min_df=2, analyzer=myTokenizer).fit(remarkers) # stop_words = ''
bow = vectorizer.transform(remarkers)

# for time
elapsed_time = time.time() - start
print(elapsed_time)

# 綺麗に書き換えられる
freqs = [(word, bow.getcol(idx).sum()) for word, idx in vectorizer.vocabulary_.items()]
order = sorted(freqs, key = lambda x: -x[1]) # tapleのlist
terms = []
nums = []
for a, b in order:
    terms.append(a)
    nums.append(b)

# 背景おかしい？
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
fig = plt.figure(figsize=(17,3))
ax = fig.add_subplot(1,1,1)
ax.set_xticklabels(terms[0:40], rotation = 90)
sns.set_style("whitegrid")

font_path = '/usr/share/fonts/truetype/takao-gothic/TakaoPGothic.ttf'
font_prop = FontProperties(fname=font_path)
matplotlib.rcParams['font.family'] = font_prop.get_name()
sns.barplot(x=terms[0:20], y=nums[0:20])

plt.title('2014の発言回数トップ20　（※人物のみ）')
```
